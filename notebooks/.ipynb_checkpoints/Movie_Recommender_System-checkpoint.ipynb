# 5- NLP Preprocessing & Metadata Feature Engineering :

- This is where: raw movie metadata becomes machine-understandable semantic information.

# 5.1- understanding Json like metadata :

- we have some columns containing data in a form : a string representation of structured JSON data. 

- we aregoing to extrat the mainiful data from it 

# 5.2- Genre Extraction : 

```python
# step 1- Creating the Extraction function :
def convert(text):
    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])
              
    return L
```

```python
# step 2- extracting the genre:
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

# 5.3- Keywords Extraction :

```python
# using the same convert function to extract keywords:
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'].head()
```

# 5.4- cast Extraction feature engineering:

- We do NOT keep ALL actors, Because: huge noise, memory explosion, weak relevance
- Real systems often keep: top-billed actors only
- for this we will only take actors with order 1 or 0 .

```python
# step 1- creating the cast extraction function 
def cast_extract(text):
    L = []

    for i in ast.literal_eval(text):
        if(i['order'] < 3):
            L.append(i['name'])

    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_extract)
```

```python
movies['cast'].head()
```

# 5.5- Director Extraction From crew:

```python
# creating the director extraction function:
def director_extrcat(text):
    L = []

    for i in ast.literal_eval(text):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(director_extrcat)
```

```python
movies['crew'].head()
```

# 5.6- Overview text processing :

- overview is the richest semantic feature in the whole dataset.
- We want a semantic content-based recommender, So we should: KEEP semantic richness and NOT over-clean.

```python
movies['overview'][0]
```

# step 1- Remove ponctuation: 

```python
def no_punc(text):
    mess = ''
    
    mess = ''.join([char for char in text if char not in string.punctuation ])

    return mess
```

```python
movies['overview'] = movies['overview'].apply(no_punc)
```

```python
movies['overview'].head(2)
```

```python
movies['overview'][0]
```

- we removed the punctuations fro the overview text. they don't have semantic meaning.

# step 2- Tokinisation :
- for this w going to use a simple split() function is enough here.

```python
# spliting on space:
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['overview'].head(2)
```

- now we have overview as tokinized word representation.

# 5.7- Removing space from multiword tokens :

```python
# Remove space from  multiwords in genre:
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies['genres'].head(2)
```

```python
# Same thng for keywords :
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
# same for cast:
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
# for crew :
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

- now we have words like james conner will be jamesconer and will be treated like one semantic entity.

# 5.8- Create Unified Semantic Tags :
- THIS is the HEART of the recommendation engine.

- Here We combine ALL semantic information into One unified semantic representation called: tags

```python
# creating the tags column :
movies['tags'] = (
    movies['overview'] +
    movies['genres'] +
    movies['keywords'] +
    movies['cast'] +
    movies['crew']
)
```

```python
movies['tags'].head(2)
```

- This is:semantic feature fusion

# 5.9- create Final NLP Dataset:

- since everything important is in the tags feature we make the new movies data set nto this :

```python
new_movies = movies[['movie_id', 'title', 'tags']]
```

```python
new_movies.head(2)
```

# 5.10- Convert tags into text :

- Currently tags are lists, and  Vectorizers need TEXT, so must convert to text.

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : " ".join(x))
```

```python
new_movies['tags'][0]
```

# 5.11- Lowercasing :
- without lowercasing Machine sees different tokens. (Action != action)

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : x.lower())
```

```python
new_movies['tags'][0]
```

# 5.12- Stemming :
Without stemming Machine sees:(love, loving, loved) as DIFFERENT words. And this is bad semantic understanding.

```python
# creating the stemmer :
ps = PorterStemmer()
```

```python
# creating the stemming function :
def stemming(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
# applying the stemming function to tags:
new_movies['tags'] = new_movies['tags'].apply(stemming)
```

```python
# result:
new_movies['tags'][0]
```
