# 5.2- Genre Extraction : 

```python
# step 1- Creating the Extraction function :
def convert(text):
    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])
              
    return L
```

```python
# step 2- extracting the genre:
movies['genres'] = movies['genres'].apply(convert)
```

# 5.3- Keywords Extraction :

```python
# using the same convert function to extract keywords:
movies['keywords'] = movies['keywords'].apply(convert)
```

# 5.4- cast Extraction feature engineering:

- We do NOT keep ALL actors, Because: huge noise, memory explosion, weak relevance
- Real systems often keep: top-billed actors only
- for this we will only take actors with order 1 or 0 .

```python
# step 1- creating the cast extraction function 
def cast_extract(text):
    L = []

    for i in ast.literal_eval(text):
        if(i['order'] < 3):
            L.append(i['name'])

    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_extract)
```

```python
movies['cast'].head()
```

# 5.5- Director Extraction From crew:

```python
# creating the director extraction function:
def director_extrcat(text):
    L = []

    for i in ast.literal_eval(text):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(director_extrcat)
```

# 5.12- Stemming :
Without stemming Machine sees:(love, loving, loved) as DIFFERENT words. And this is bad semantic understanding.

```python
# creating the stemmer :
ps = PorterStemmer()
```

```python
# creating the stemming function :
def stemming(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
# applying the stemming function to tags:
new_movies['tags'] = new_movies['tags'].apply(stemming)
```

Now we will transform semantic text into mathematical vectors so machines can:

- compare movies

# in this first version we will use the Bag-of-words method :

- bag of words It represents text using: word frequency, ignoring grammar, order and deep context.

# 6.1- Using CountVevtorizer: 

```python
movies_cv = CountVectorizer(max_features=5000, stop_words='english')
```

- Keeps only top 5000 most important words. Without limit: vocabulary explode, memory usage increases, noisy rare words appear.
- also Earlier we kept StopWrods for semantic preservation, but now: CountVectorizer can remove them more intelligently.

```python
# create the vectors:
vectors = movies_cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
# checkng the vectors shape:
vectors.shape
```

- meaning 4806 movies, and 5000 word vocabulary

# 6.2- Similarity Modeling:
- Movies are similar if: their vectors point in similar directions
- We measure this mathematically.
- there a lot method to calculate the distance, but we are going to use the cosine similarity (best for NLP ) because orientation matters more than magnitude.

```python
# Create the smiliraty matrix : using the Cosine similarity method : 
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

- Every movie compared to every other movie. so we have a matrix with all movies in row with all movies in columns , and the value are the similarit between them (value btween 0 and 1)

# 6.3- Creating the Recomender system function :

```python
def recommend(movie):

    #getting the movie index:
    movie_index = new_movies[new_movies['title'] == movie].index[0]

    #getting the similarity score for this movie index:
    distances = similarity[movie_index]

    # adding movies inces with enumerated the sorting them Descending on the similarity score using python list:
    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x:x[1]
    )[1:6]

    # after removing it self staring from 1 in the list, we loop to get their top 5 recomendetion titles.
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
# testing the recomender function :
recommend('Avatar')
```

- The Bag of Words model produced highly intuitive and genre-consistent recommendations. Most suggested movies strongly match Avatar through explicit sci-fi and alien-war related keywords.
- This version performs well because Avatar contains many distinctive sci-fi terms that also appear directly in these movies.
However, the recommendations are still based mainly on: surface-level word matching

# 7.3- Building the TFIDF Recomender function :

```python
def recommend_tfidf(movie):

    movie_index = new_movies[new_movies['title'] == movie].index[0]

    distances = tfidf_similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
# testing the TFIDF Recomender:
recommend_tfidf('Avatar')
```

- The TF-IDF system produced slightly more refined recommendations compared to Bag of Words. Because TF-IDF reduces the importance of extremely common words, the model focused more on distinctive and informative sci-fi terms.

- This improved the recommendation quality by reducing noisy generic words.

Positive observations:
- Star Trek Into Darkness is a stronger semantic sci-fi match than some Bag of Words outputs
- less dependence on overly common terms
-  more importance given to unique contextual vocabulary

However, some recommendations reveal TF-IDF limitations:
- Meet Dave appears less semantically aligned with Avatar despite sharing some space/alien-related vocabulary.

TF-IDF still lacks deep contextual understanding

# 8.3- Training the Word2Vec Model :

```python
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

We used: 
- our sentences as the curpos to train the model .
- each word becomes 100-dementional semantic vector (dense vectors).
- context window size of 5 , to capture semantic meaning of 5 word on each side .
- we told the model to ignore rare words appearing fewer than 2 times.
- we use 4 parallel cpu to train faster.

```python
# cheking the vocabulary size:
len(w2v_model.wv.index_to_key)
```

- the model learned 15079 words from the movies sentences corpus used inside the model.

```python
# lets undettand each word how it represented in a dense vecotr:
w2v_model.wv['alien']
```

- this is the 100-dementional sematic dense vector for the word 'alien' for exemple.This is semantic embedding.

```python
# let's also check for similar words to alien :
w2v_model.wv.most_similar('alien')
```

- this proves that semantic learning happend, and this works because : words used in similar contexts become nearby vectors.
This is distributional semantics.

# 8.4- Movie Embedding :

- Word2Vec gives words embbedings, but we need movie embedding to create vectors for each movie so we can compre it to each other not just words.
- to do this we are going to average all word vectors inside each movie. to create the movie vector.

```python
# creating the movie embedding function :
def document_vector(doc):

    # chekcing if the movie words has its embedding vector in our model leanred vocab.
    words = [word for word in doc if word in w2v_model.wv]

    # if all words dont exist we create the 100-dementional vector with all zeros.
    if len(words) == 0:
        return np.zeros(100)

    # returning the movie vector as average of all words vector inside it.
    return np.mean(w2v_model.wv[words], axis=0)
```

```python
# now using the function to ceating the movies embeddings:
movie_vectors = np.array(new_movies['tokens'].apply(document_vector).tolist())
```

```python
# checking the shape of it :
movie_vectors.shape
```

- matrix with all movies in row each row a single semantic dense vector of 100 demention 

# 8.5- Modeling the Similarity with Cosine similarity :

```python
w2v_similarity = cosine_similarity(movie_vectors)
```

- same logic because the same recommender pipline , only the vector representation that change:

```python
w2v_similarity.shape
```

# 8.6- Creating the recommander function :

```python
def recommend_w2v(movie):

    movie_index = new_movies[new_movies['title'] == movie].index[0]

    distances = w2v_similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x:x[1]
    )[1:6]

    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
# testing the Word2Vec recomender :
recommend_w2v('Avatar')
```

- This result is extremely important because it reveals both the power and limitations of basic Word2Vec document averaging.

The recommendations became much less keyword-driven and more influenced by learned semantic vector relationships. This proves the model moved beyond simple word overlap.

However, the results also became noticeably unstable and less relevant to Avatar.

But major issues appeared:
